```python
import matplotlib.dates as mdates


df['date of reservation'] = pd.to_datetime(df['date of reservation'], format='mixed', dayfirst=False, errors='coerce')

# Extract the date part
df['booking_date'] = df['date of reservation'].dt.date

# Count occurrences of each booking date
booking_counts = df['booking_date'].value_counts()
booking_counts = booking_counts.sort_values(ascending=False)

most_booked_date = booking_counts.index[0]

# Convert booking_counts.index to matplotlib dates
booking_dates_num = mdates.date2num(booking_counts.index)

all_dates = booking_counts.index
dates_for_labels = all_dates[::120]
dates_for_labels_num = mdates.date2num(dates_for_labels)

# Create the plot
plt.figure(figsize=(15, 6))
plt.bar(booking_dates_num, booking_counts.values)  # Use booking_dates_num here
plt.xlabel('Booking Date')
plt.ylabel('Number of Bookings')
plt.title('Most Booked Date')

date_format = mdates.DateFormatter('%Y-%m')  # Choose your desired date format
plt.gca().xaxis.set_major_formatter(date_format)
plt.xticks(rotation=45, ha='right')

# # Highlight the most booked date
plt.axhline(y=booking_counts[0], color='red', linestyle='--', label=f'Most Booked Date: {most_booked_date}')
plt.legend()

plt.show()

print(f"The most booked date is: {most_booked_date} with {booking_counts.values[0]} bookings")
```

## Removing Outliers

```python
# prompt: remove the ouliers in the average price column

# Calculate Q1, Q3, and IQR
Q1 = df['average price'].quantile(0.25)
Q3 = df['average price'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_no_outliers = df[(df['average price'] >= lower_bound) & (df['average price'] <= upper_bound)]

# Print the number of rows before and after removing outliers
print(f"Number of rows before removing outliers: {len(df)}")
print(f"Number of rows after removing outliers: {len(df_no_outliers)}")
```

```python
print(f"Number of rows before removing outliers: {len(df_no_outliers)}")
max_lead_time = df_no_outliers['lead time'].quantile(0.95) # Example: using 95th percentile

# Replace lead times above the threshold with the maximum non-outlier value
df_no_outliers.loc[df_no_outliers['lead time'] > max_lead_time, 'lead time'] = max_lead_time
print(f"Number of rows after removing outliers: {len(df_no_outliers)}")
```

```python
# prompt: remove the bookings with 0 adults

# Remove bookings with 0 adults
df_no_outliers = df_no_outliers[df_no_outliers['number of adults'] != 0]

print(f"Number of rows after removing bookings with 0 adults: {len(df_no_outliers)}")

df_no_outliers = df_no_outliers[df_no_outliers['number of children'] < 9]

print(f"Number of rows after removing bookings with 9 and 10 children: {len(df_no_outliers)}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Get unique market segments
market_segments = df_no_outliers['market segment type'].unique()
num_segments = len(market_segments)

# Create a 3x2 grid, then remove unwanted subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 12))  # 3 rows, 2 columns

# Remove unwanted subplots (3,0) and (3,2)
fig.delaxes(axes[2, 0])  # Remove subplot at (3, 0)
fig.delaxes(axes[2, 1])  # Remove subplot at (3, 2)

# Iterate through market segments and plot pie charts
for i, segment in enumerate(market_segments):
    if i < 4:  # First 4 segments in 2x2 grid
        row = i // 2
        col = i % 2
        ax = axes[row, col]
    else:  # Last segment in bottom middle
        ax = fig.add_subplot(3, 2, 5)  # Add subplot at position 5
    segment_data = df_no_outliers[df_no_outliers['market segment type'] == segment]
    booking_status_counts = segment_data['booking status'].value_counts()
    ax.pie(booking_status_counts, labels=booking_status_counts.index, startangle=90, colors=['skyblue', 'lightcoral'])
    ax.set_title(f'Market Segment: {segment}')

plt.show()
```
