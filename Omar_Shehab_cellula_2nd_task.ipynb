```python
# Convert date and extract features
df['date of reservation'] = pd.to_datetime(df['date of reservation'],errors='coerce')
df['year of reservation'] = df['date of reservation'].dt.year
df['month of reservation'] = df['date of reservation'].dt.month
df['day of reservation'] = df['date of reservation'].dt.day
```

```python
df.head()
```

```python
df.info()
```

```python

# Handle outliers (capping at 1st and 99th percentile)
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    lower_bound = df[col].quantile(0.01)
    upper_bound = df[col].quantile(0.99)
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
df['type of meal']= label_encoder.fit_transform(df['type of meal'])
df['market segment type']= label_encoder.fit_transform(df['market segment type'])
df['room type']= label_encoder.fit_transform(df['room type'])
df['booking status']= label_encoder.fit_transform(df['booking status'])
df['type of meal'].unique()
```
